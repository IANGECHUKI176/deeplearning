```python
import torch
import torch.nn as nn
from torchvision import datasets , transforms
from torchsummary import summary
from torch.utils.data import DataLoader
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
import itertools
from tqdm import tqdm
```

```python
n_epochs = 200
batch_size = 64
lr = 0.0002
b1 = 0.5
b2 = 0.999
latent_dim = 62
code_dim = 2
n_classes = 10
img_size = 32
channels = 1
display_step = 5000
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
def weights_init_normal(m : nn.Module):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        torch.nn.init.normal_(m.weight.data,0.0,0.02)
    elif classname.find("BatchNorm2d") != -1:
        torch.nn.init.normal_(m.weight.data,1.0,0.02)
        torch.nn.init.constant_(m.bias.data,0.0)

```

```python
def to_categorical(y,num_columns):
    """ Return one hot encoded variable"""
    y_cat = torch.zeros((y.size(0),num_columns)).to(device)
    y_cat[range(y.size(0)),y] = 1
    return y_cat.to(device)
```

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator,self).__init__()

        input_dim = latent_dim + n_classes + code_dim

        self.init_size = img_size // 4 #initial size b4 upsampling
        self.l1 = nn.Linear(input_dim , 128 * self.init_size ** 2)

        self.conv_blocks = nn.Sequential(
            nn.BatchNorm2d(128),
            nn.Upsample(scale_factor = 2),
            nn.Conv2d(128,128,3,1,1),
            nn.BatchNorm2d(128,0.8),
            nn.LeakyReLU(0.2,inplace = True),
            nn.Upsample(scale_factor = 2),
            nn.Conv2d(128,64,3,1,1),
            nn.BatchNorm2d(64,0.8),
            nn.LeakyReLU(0.2,inplace = True),
            nn.Conv2d(64,channels,3,1,1),
            nn.Tanh()
        )
    def forward(self,noise,labels,code):
        gen_input = torch.cat((noise ,labels ,code),-1)
        out = self.l1(gen_input)
        out = out.view(out.size(0),128,self.init_size,self.init_size)
        img = self.conv_blocks(out)
        return img
```

```python
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator,self).__init__()

        def discriminator_block(in_features,out_features,bn = True):
            block = [nn.Conv2d(in_features,out_features,3,2,1),nn.LeakyReLU(0.2,inplace = True),nn.Dropout(0.25)]
            if bn:
                block.append(nn.BatchNorm2d(out_features,0.8))
            return block
        self.conv_blocks = nn.Sequential(
            *discriminator_block(channels,16,bn = False),
            *discriminator_block(16,32),
            *discriminator_block(32,64),
            *discriminator_block(64,128)
        )
        #width and height of downsampled image
        ds_size = img_size // 2**4

        #output layers
        self.adv_layer = nn.Sequential(nn.Linear(128*ds_size**2,1))
        self.aux_layer = nn.Sequential(nn.Linear(128*ds_size**2,n_classes),nn.Softmax())
        self.latent_layer = nn.Sequential(nn.Linear(128*ds_size**2,code_dim))
    def forward(self,img):
        out = self.conv_blocks(img)
        out = out.view(out.size(0),-1)
        validity = self.adv_layer(out)
        label = self.aux_layer(out)
        latent_code = self.latent_layer(out)
        return validity,label,latent_code
```

```python
generator = Generator().to(device)
discriminator = Discriminator().to(device)

#loss weights
lambda_cat = 1.0
lambda_con = 0.1

#init weights
generator.apply(weights_init_normal)
discriminator.apply(weights_init_normal)

#optimizer
optimizer_G = torch.optim.Adam(generator.parameters(), lr = lr,betas = (b1,b2))
optimizer_D = torch.optim.Adam(discriminator.parameters(),lr = lr,betas = (b1,b2))
optimizer_info = torch.optim.Adam(itertools. chain(generator.parameters(),discriminator.parameters()),lr = lr,betas = (b1,b2))

#loss functions
adversarial_loss = nn.MSELoss()
categorical_loss = nn.CrossEntropyLoss()
continuous_loss = nn.MSELoss()
```

dataloader

```python
dataloader = DataLoader(
    datasets.MNIST(
        './root',
        download = True,
        train = True,
        transform = transforms.Compose([
            transforms.Resize(img_size),
            transforms.ToTensor(),
            transforms.Normalize([0.5],[0.5])
        ])
    ),
    batch_size = batch_size,
    shuffle = True
)
```

```python
def show_tensor_images(image_tensor, num_images=25, size=(1, 28, 28)):
    '''
    Function for visualizing images: Given a tensor of images, number of images, and
    size per image, plots and prints the images in an uniform grid.
    '''
    image_tensor = (image_tensor + 1) / 2
    image_unflat = image_tensor.detach().cpu()
    image_grid = make_grid(image_unflat[:num_images], nrow=5)
    plt.imshow(image_grid.permute(1, 2, 0).squeeze())
    plt.show()
```

```python
images,labels = next(iter(dataloader))
show_tensor_images(images)
```

```python
import torch.nn.functional as F

def show_ordered_grid_of_generated_images(generator, n_row, n_classes, latent_dim, code_dim, device):
    """
    Displays an ordered grid of generated images with each row containing images of the same class.
    """
    images_per_class = n_row
    n_samples = images_per_class * n_classes

    z = torch.randn(n_samples, latent_dim).to(device)
    label_input = F.one_hot(torch.arange(n_classes).repeat(images_per_class), num_classes=n_classes).to(device)
    code_input = torch.tensor(np.random.uniform(-1, 1, (n_samples, code_dim)), dtype=torch.float32).to(device)

    gen_imgs = generator(z, label_input.float(), code_input)

    image_tensor = (gen_imgs + 1) / 2
    image_unflat = image_tensor.detach().cpu()

    rows = []
    for class_idx in range(n_classes):
        row_images = image_unflat[class_idx * images_per_class : (class_idx + 1) * images_per_class]
        rows.append(torch.cat([img.unsqueeze(0) for img in row_images], dim=0))

    ordered_grid = torch.cat(rows, dim=0)

    plt.figure(figsize=(5, 5))  # Adjust the figure size if needed
    grid = make_grid(ordered_grid, nrow=images_per_class, normalize=True, scale_each=True)
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis('off')

    plt.show()
```

```python
cur_step = 0
mean_generator_loss = 0
mean_discriminator_loss = 0
mean_info_loss = 0
for epoch in range(n_epochs):
    for i,(imgs,labels) in enumerate(tqdm(dataloader)):
        real_imgs , labels = imgs.to(device), labels.to(device)

        batch_size = real_imgs.size(0)

        valid = torch.ones(batch_size,1).to(device)
        fake = torch.zeros(batch_size,1).to(device)

        #congifure input
        real_imgs = real_imgs.requires_grad_(True)
        #labels = to_categorical(la)

        # -----------------
        #  Train Generator
        # -----------------
        optimizer_G.zero_grad()
        z = torch.randn(batch_size,latent_dim).to(device)
        label_input = to_categorical(torch.randint(0,n_classes,(batch_size,)),num_columns = n_classes)
        code_input = torch.tensor(np.random.uniform(-1, 1, (batch_size, code_dim)), dtype=torch.float32).to(device)

        #generate a batch of images
        gen_imgs = generator(z,label_input,code_input)

        #loss measures generator's ability to fool discriminator
        validity,_,_ = discriminator(gen_imgs)
        g_loss = adversarial_loss(validity,valid)

        g_loss.backward()
        optimizer_G.step()

        # ---------------------
        #  Train Discriminator
        # ---------------------
        optimizer_D.zero_grad()
        #loss for real images
        real_pred,_,_ = discriminator(real_imgs)
        d_real_loss = adversarial_loss(real_pred,valid)

        #loss for fake images
        fake_pred,_,_ = discriminator(gen_imgs.detach())
        d_fake_loss = adversarial_loss(fake_pred,fake)

        #total discrminator loss
        d_loss = (d_real_loss + d_fake_loss) / 2

        d_loss.backward()
        optimizer_D.step()

        # ------------------
        # Information Loss
        # ------------------
        optimizer_info.zero_grad()

        # Sample labels
        sampled_labels = np.random.randint(0, n_classes, batch_size)
        # Ground truth labels
        gt_labels = torch.tensor(sampled_labels).requires_grad_(False).to(device)

        # Sample noise, labels and code as generator input
        z = torch.randn(batch_size,latent_dim).to(device)
        label_input = to_categorical(gt_labels,num_columns=n_classes)
        code_input = torch.tensor(np.random.uniform(-1, 1, (batch_size, code_dim)), dtype=torch.float32).to(device)

        gen_inputs = generator(z,label_input,code_input)
        _,pred_labels,pred_code = discriminator(gen_inputs)

        info_loss = lambda_cat * categorical_loss(pred_labels,gt_labels) + lambda_con * continuous_loss(pred_code,code_input)

        info_loss.backward()
        optimizer_info.step()
        # --------------
        # Log Progress
        # --------------


        # Keep track of the average discriminator loss
        mean_discriminator_loss += d_loss.item() / display_step

        # Keep track of the average generator loss
        mean_generator_loss += g_loss.item() / display_step

        mean_info_loss += info_loss.item() / display_step
        # After each epoch, visualize the generated images
        if cur_step % display_step == 0 and cur_step > 0:
            print(f"Epoch {epoch}, step {cur_step}: G loss: {mean_generator_loss}, d loss: {mean_discriminator_loss},info loss: {mean_info_loss}")
            print('----------------gen images-------------')
            show_ordered_grid_of_generated_images(generator,n_row = 5,n_classes = n_classes,latent_dim = latent_dim,code_dim = code_dim,device = device)

            mean_generator_loss = 0
            mean_discriminator_loss = 0
        cur_step += 1
```
